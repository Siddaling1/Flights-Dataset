```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv(r"/content/airlines_flights_data.csv")
```

```python
data
```

```python
data.drop(columns= "index",inplace=True)
```

# **2.what are the frequency of airlines?**

```python
data['airline'].value_counts()
```

```python
data['airline'].value_counts().plot.bar(color=["red"])
plt.title ("Airline with Frequency")
plt.xlabel("Airline")
plt.ylabel("Frequency")
plt.show()
```

# **3. Graph of departure_time and arival_time by frequency?**

```python
data["departure_time"].value_counts()
```

```python
data["departure_time"].value_counts().plot.bar(color=["green"])
plt.title("Departure Time by Frequency")
plt.xlabel("Departure Time")
plt.ylabel("Frequency")
plt.show()
```

```python
data["arrival_time"].value_counts()
```

```python
data["arrival_time"].value_counts().plot.bar(color=["green"])
plt.title("Arrival Time by Frequency")
plt.xlabel("Arriaval Time")
plt.ylabel("Frequency")
plt.show()
```

# **4.Show bar grapy of source city and destination city by frequency?**

```python
data["source_city"].value_counts()
```

```python
data["source_city"].value_counts().plot.barh(color=["blue"])
plt.title("Source City by Frequency")
plt.xlabel("Source City")
plt.ylabel("Frequency")
plt.show()
```

```python
data['destination_city'].value_counts()
```

```python
data['destination_city'].value_counts().plot.barh(color=["blue"])
plt.title("Destination city by Frequency")
plt.xlabel("Destination city")
plt.ylabel("Frequency")
plt.show()
```

# **5.Does price varies with airlines?**

```python
data.head()
```

```python
data.groupby("airline")["price"].mean()
```

```python
sns.catplot( x = 'airline', y = 'price', kind = 'bar', palette = 'rocket', data = data, hue = 'class')

plt.show()
```

# **6.Does price changes for Destion time and arrival time?**

```python
data.head()
```

```python
data.groupby('departure_time')['price'].mean()
```

```python
sns.catplot( x = 'departure_time', y = 'price', kind = 'bar', data = data )

plt.show()
```

```python
data.groupby('arrival_time')['price'].mean()
```

```python
sns.catplot( x = 'arrival_time', y = 'price', kind = 'bar', data = data )

plt.show()
```
